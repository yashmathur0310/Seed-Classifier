```python
import os 
```

```python
categories=os.listdir('train')
```

```python
folders=['train','test']
```

```python
for folder in folders:
    for category in categories:
        print(f"{folder} {category} : {len(os.listdir(folder+'/'+category))}")
    print('-----------------------')
```

```python
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
import random
def random_image_display(filename,category):
    path=filename+'/'+category
    img=random.sample(os.listdir(path),1)
    img=mpimg.imread(path+'/'+img[0])
    plt.imshow(img)
    print(img.shape)
```

```python
random_image_display('train','bad-sunflower')
```

```python
random_image_display('test','good-sunflower')
```
